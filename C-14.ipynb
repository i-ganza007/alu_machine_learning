```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 28651, 5730)
r = np.log(0.5)
t = 5730
fig, ax = plt.subplots() 
y = np.exp((r / t) * x)
plt.xticks(np.arange(0,26000,5000))
plt.yticks = [10**(-i) for i in range(0, 6)]
plt.plot(x,y,c='blue')
ax.set_yscale('log')
plt.show()
```
