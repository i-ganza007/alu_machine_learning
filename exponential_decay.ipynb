```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,21000,1000)
r = np.log(0.5)
t1 = 5730
t2 = 1600
y1 = [np.exp((r/t1)*i)for i in x]
y2 = [np.exp((r/t2)*y) for y in x]
plt.xticks(np.arange(0,21000,2500))
plt.yticks(np.arange(0,max(y1)+0.2,0.2))
plt.title('Exponential Decay of Radioactive Elements')
plt.xlabel('Time (years)')
plt.ylabel('Fraction Remaining')
plt.plot(x,y1,c='red',linestyle='--',label='C-14')
plt.plot(x,y2,c='green',linestyle='-',label='Ra-226')
plt.legend()
plt.show()
```
